```python
import requests
from bs4 import BeautifulSoup

# 요청할 URL
url = "https://www.carwiki.co.kr/brandList"

# 웹 페이지 HTML 요청
response = requests.get(url)
response.raise_for_status()  # 요청 에러 시 중단

# BeautifulSoup 객체 생성
soup = BeautifulSoup(response.text, 'html.parser')

# 데이터를 저장할 리스트
data = []

# li 태그를 반복하며 이미지와 이름 크롤링
for li in soup.find_all('li'):
    img_tag = li.find('img')  # img 태그 찾기
    span_tag = li.find('span')  # span 태그 찾기

    # img src와 span 텍스트 추출
    img_src = img_tag['src'] if img_tag else None
    span_text = span_tag.text.strip() if span_tag else None

    # 데이터를 리스트에 추가
    if img_src and span_text:
        data.append({'img_src': img_src, 'name': span_text})

# 크롤링한 데이터 출력
for item in data:
    print(f"이미지 URL: {item['img_src']}, 이름: {item['name']}")
```
